```python
from pyspark import SparkContext
sc = SparkContext("local", "RDD Demo")
```

```python
data = [1, 2, 3, 4, 5]
rdd = sc.parallelize(data)
```

```python
print(sc)
```

```python
rdd
```

```python
rdd.take(3)
```

```python
rdd.count()
```

```python
rdd_squared = rdd.map(lambda x: x**2)
rdd_squared.collect()
```

```python
rdd_even = rdd.filter(lambda x: x % 2 == 0)
rdd_even.collect()
```

```python
rdd_words = sc.parallelize(["hello world", "spark rdd"])
rdd_words_flat = rdd_words.flatMap(lambda x: x.split(" "))
rdd_words_flat.collect()
```

```python
rdd_distinct = sc.x([1, 2, 2, 3, 4, 4, 5]).distinct()
rdd_distinct.collect()
```

```python
rdd_distinct.reduce(lambda a, b: a + b)
```

```python
rdd_pairs = sc.parallelize([("a", 1), ("b", 2), ("a", 3), ("b", 4)])
rdd_sum = rdd_pairs.reduceByKey(lambda x, y: x + y)
rdd_sum.collect()
```

```python
rdd_grouped = rdd_pairs.groupByKey()
[(k, list(v)) for k, v in rdd_grouped.collect()]
```

```python
rdd_sorted = rdd_distinct.sortBy(lambda x: x, ascending=False)
rdd_sorted.collect()
```

```python
rdd_distinct.sample(withReplacement=False, fraction=0.5).collect()
```

```python
rdd1 = sc.parallelize([("a", 1), ("b", 2)])
rdd2 = sc.parallelize([("a", 3), ("b", 4)])
rdd_joined = rdd1.join(rdd2)
rdd_joined.collect()
```
